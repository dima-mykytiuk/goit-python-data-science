```python
from scipy.integrate import odeint 
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.5
b = 0.3
N = 1000000
S0 = 990000
I0 = 7000
R0 = 3000
t0 = 0
tf = 25
t = np.linspace(t0, tf, N)
```

```python
def deriv(y, t, N, a, b):
    S, I, R = y
    dSdt = -a * S
    dIdt = a * S - b * I
    dRdt = b * I
    return dSdt, dIdt, dRdt


y0 = S0, I0, R0

ret = odeint(deriv, y0, t, args=(N, a, b))
S, I, R = ret.T
```

Нахождение R(t) по формуле S(t) + I(t) + R(t) = N. Не знаю зачем оно надо если функция сверху все и так делает.

```python
tick = 0
res = []
for _ in range(len(S)):
    R_2 = N - (S[tick] + I[tick])
    res.append(R_2)
    tick += 1
```

```python
plt.plot(t, S, 'b', label = 'Sensetive')
plt.title('S(t)',size = 20)
plt.xlabel('Time(t)') 
plt.ylabel('Number(N)') 
plt.legend()
plt.grid() 
plt.show()
```

```python
plt.plot(t, I, 'b', label = 'Sensetive')
plt.title('I(t)',size = 20)
plt.xlabel('Time(t)') 
plt.ylabel('Number(N)') 
plt.legend()
plt.grid() 
plt.show()
```

```python
plt.plot(t, R, 'b', label = 'Sensetive')
plt.title('R(t)',size = 20)
plt.xlabel('Time(t)') 
plt.ylabel('Number(N)') 
plt.legend()
plt.grid() 
plt.show()
```

```python
plt.plot(t, S, 'b', label = 'Sensetive')
plt.plot(t, I, 'r', label = 'Infected')
plt.plot(t, R, 'g', label = 'Recovered')
plt.title('Final result',size = 20)
plt.legend()
plt.grid() 
plt.show()
```
