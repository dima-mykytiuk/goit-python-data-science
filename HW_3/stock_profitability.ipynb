```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
income_df = pd.read_excel('stock_income.xlsx')
```

```python
income_df
```

```python
sns.lmplot(x="Company_A",
           y="Company_B",
           data=income_df,
           scatter_kws={'s': 150, 'edgecolor': 'black'},
           line_kws={'lw':5, 'ls':':'}
          )
plt.title('Correlation plot between A and B', fontsize=15)
plt.grid()
plt.show()
```

```python
sns.lmplot(x="Company_A",
           y="Company_C",
           data=income_df,
           scatter_kws={'s': 150, 'edgecolor': 'black'},
           line_kws={'lw':5, 'ls':':'}
          )
plt.title('Correlation plot between A and C', fontsize=15)
plt.grid()
plt.show()
```

```python
sns.lmplot(x="Company_B",
           y="Company_C",
           data=income_df,
           scatter_kws={'s': 150, 'edgecolor': 'black'},
           line_kws={'lw':5, 'ls':':'}
          )
plt.title('Correlation plot between B and C', fontsize=15)
plt.grid()
plt.show()
```
