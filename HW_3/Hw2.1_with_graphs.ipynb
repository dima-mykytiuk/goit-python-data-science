```python
import pandas as pd
import requests
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Прочитайте данные с помощью метода read_html из таблицы "Коефіцієнт народжуваності в регіонах України (1950—2019)" по ссылке

```python
url = 'https://uk.wikipedia.org/wiki/Населення_України#Народжуваність'
r = requests.get(url)
website = r.text
table = pd.read_html(website, match='Коефіцієнт народжуваності в регіонах України', encoding="UTF-8", thousands='.', decimal=',')
```

```python
print(table[0])
```

```python
df_birth_rate = pd.DataFrame(table[0])
```

# Вывести первые строки таблицы с помощью метода head

```python
df_birth_rate.head()
```

# Определите количество строк и столбцов в датафрейме (атрибут shape)

```python
df_birth_rate.shape
```

# Замените в таблице значения "—" на значения NaN

```python
df_birth_rate.replace('—', np.nan, inplace=True)
```

```python
df_birth_rate
```

# Определите типы всех столбцов с помощью dataframe.dtypes

```python
df_birth_rate.dtypes
```

# Замените типы не числовых колонок на числовые. Подсказка - это колонки где находился символ "—"

```python
df_birth_rate['2014'] = df_birth_rate['2014'].astype('float')
df_birth_rate['2019'] = df_birth_rate['2019'].astype('float')
df_birth_rate.dtypes
```

# Посчитайте, какая доля пропусков содержится в каждой колонке (используйте методы isnull и sum)

```python
df_birth_rate.isnull().sum()
```

# Удалите из таблицы данные по всей стране, последняя строчка таблицы

```python
df_birth_rate.drop(index=df_birth_rate.index[-1], axis=0, inplace=True)
```

```python
df_birth_rate
```

# Замените отсутствующие данные в столбцах средними значениями по этим столбцам (метод fillna)

```python
for item in df_birth_rate.keys():
    if item.isdigit():
        df_birth_rate.fillna({item: df_birth_rate[item].mean()}, inplace=True)
```

```python
df_birth_rate
```

# Получите список регионов, где уровень рождаемости в 2019 году был выше среднего по Украине

```python
df_birth_rate[['Регіон', '2019']].where(df_birth_rate['2019'] > df_birth_rate['2019'].mean()+0.00001).dropna()
```

# В каком регионе была самая высокая рождаемость в 2014 году?

```python
max_birth_region = df_birth_rate['2014'].idxmax()
df_birth_rate['Регіон'][max_birth_region]
```

# Постройте столбцовую диаграмму рождаемости по регионам в 2019 году

```python
df = pd.DataFrame({'Регіон':df_birth_rate['Регіон'], 'Коефіцієнт':df_birth_rate['2019']})
ax = df.plot.bar(x='Регіон',
                 y='Коефіцієнт',
                 figsize=(10,5),
                 xlabel='Регіон',
                 ylabel="Коефіцієнт",
                 title="Коефіцієнт народжуваності в регіонах України за 2019 рік",
                 width=0.9,
                 color='red',
                )
```

```python
df_1 = (df_birth_rate['Регіон'], df_birth_rate['1950'])
df_2 = (df_birth_rate['Регіон'], df_birth_rate['1960'])
df_3 = (df_birth_rate['Регіон'], df_birth_rate['1970'])
df_4 = (df_birth_rate['Регіон'], df_birth_rate['1990'])
plt.figure(figsize=(15, 8))
plt.title('Рождаемость в Украине в составе СССР 1950-1990', fontsize=15)
plt.xlabel('Город', fontsize='medium', color='midnightblue')
plt.ylabel('Рождаемость', fontsize='medium', color='midnightblue')
plt.plot(df_1[0], df_1[1], label='1950', marker="D")
plt.plot(df_2[0], df_2[1], label='1960', marker="D")
plt.plot(df_3[0], df_3[1], label='1970', marker="D")
plt.plot(df_4[0], df_4[1], label='1990', marker="D")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
df_1 = (df_birth_rate['Регіон'], df_birth_rate['2000'])
df_2 = (df_birth_rate['Регіон'], df_birth_rate['2012'])
df_3 = (df_birth_rate['Регіон'], df_birth_rate['2014'])
df_4 = (df_birth_rate['Регіон'], df_birth_rate['2019'])
plt.figure(figsize=(15, 8))
plt.title('Рождаемость в независимой Украине 2000-2019', fontsize=15)
plt.xlabel('Город', fontsize='medium', color='midnightblue')
plt.ylabel('Рождаемость', fontsize='medium', color='midnightblue')
plt.plot(df_1[0], df_1[1], label='2000', marker="D")
plt.plot(df_2[0], df_2[1], label='2012', marker="D")
plt.plot(df_3[0], df_3[1], label='2014', marker="D")
plt.plot(df_4[0], df_4[1], label='2019', marker="D")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.regplot(x ='1990', y = '2014', color='b', data = df_birth_rate)
plt.title('График корреляции между рождаемостью в 1990 и 2014 гг.', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df_birth_rate.corr(), xticklabels=df_birth_rate.corr().columns, yticklabels=df_birth_rate.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Корреляции между всеми возможными парами', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.yticks(rotation=360)
plt.show()
```
