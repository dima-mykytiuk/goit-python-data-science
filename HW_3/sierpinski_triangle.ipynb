```python
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sierpinski_triangle(coords):
    # Вариант для 1 точки внутри треугольника который я нашел на stackoverflow:
    x, y = sorted([random.random(), random.random()])
    s = x
    t = y - x
    u = 1 - y
    coord = (s * coords[0][0] + t * coords[1][0] + u * coords[2][0],
             s * coords[0][1] + t * coords[1][1] + u * coords[2][1])
    # Альтернативный вариант взять просто вершину B и по ее линии к Х взять рандом координату.
    #coord = coords[1][0], random.uniform(coords[1][0], coords[1][1])
    coord_to_display = []
    for _ in range(3000):
        random_coord = random.choice(coords)
        coord_x = random_coord[0]
        coord_y = random_coord[1]
        find_coord = ((coord[0] + coord_x) / 2, (coord[1] + coord_y) / 2)
        coord = find_coord
        coord_to_display.append(find_coord)

    x = [1.0, 2.0, 4.0, 1.0] + [x[0] for x in coord_to_display]
    y = [1.0, 4.0, 2.0, 1.0] + [y[1] for y in coord_to_display]
    plt.scatter(x, y)
```

```python
sierpinski_triangle([(1.0, 1.0), (2.0, 4.0), (4.0, 2.0)])
```
